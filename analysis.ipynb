```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
%matplotlib inline

source1 = pd.read_csv('log-100-12.csv', index_col=0, header=None).transpose()
source2 = pd.read_csv('log-100-24.csv', index_col=0, header=None).transpose()
source3 = pd.read_csv('log-100-60.csv', index_col=0, header=None).transpose()
```

```python
distance1 = source1['distance'].tolist()
distance2 = source2['distance'].tolist()
distance3 = source3['distance'].tolist()
a = np.array(list(range(0,len(distance1))))
x = a.astype(float)

plt.plot(x, distance1, 'o', label='w=12')
plt.plot(x, distance2, 'o', label='w=24')
plt.plot(x, distance3, 'o', label='w=36')
plt.legend()
plt.show()
```

```python
uniques1 = source1['unique'].tolist()
uniques2 = source2['unique'].tolist()
uniques3 = source3['unique'].tolist()
a = np.array(list(range(0,len(uniques1))))
x = a.astype(float)

plt.plot(x, uniques1, 'o', label='n=100')
plt.plot(x, uniques2, 'o', label='n=400')
plt.plot(x, uniques3, 'o', label='n=1000')
plt.legend()
plt.show()
```
